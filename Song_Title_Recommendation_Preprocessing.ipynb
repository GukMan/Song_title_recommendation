```python
import pandas as pd

# concat
path1='/content/drive/MyDrive/nlp_project/전처리/df_title_lyris_0_1000.csv'
path2='/content/drive/MyDrive/nlp_project/전처리/df_title_lyris_2000_3000.csv'
path3='/content/drive/MyDrive/nlp_project/전처리/df_title_lyris_3000_4000.csv'
df_title_lyris1= pd.read_csv(path1, encoding='utf-8-sig')
df_title_lyris2= pd.read_csv(path2, encoding='utf-8-sig')
df_title_lyris3= pd.read_csv(path3, encoding='utf-8-sig')

```

```python
df_title_lyris3
```

```python
import re 

def preprocess_kor_eng_sentence(sent):
  # 단어와 구두점 사이에 공백을 만듭니다.
  # Ex) "he is a boy." => "he is a boy ."
  sent = re.sub(r"([?.!,¿])", r" \1", sent)

  # (a-z, A-Z, ".", "?", "!", ",") 이들을 제외하고는 전부 공백으로 변환합니다.
  sent = re.sub(r"[^a-zA-Z가-힣ㄱ-ㅎㅏ-ㅣ!.?]+", r" ", sent)

  # 다수 개의 공백을 하나의 공백으로 치환
  sent = re.sub(r"\s+", " ", sent)
  
  # br 없애기
  sent = re.sub('br','',sent)

  #sent = okt.morphs(sent, stem = False)

  return sent

def preprocessing_lyris_list(list):
  list_lyris_re=[]
  for index,lyris in enumerate(list):
    list_lyris_re.append(preprocess_kor_eng_sentence(lyris))
  return list_lyris_re

def df_title_lyris_preprocessing(df):
  list_lyris_re=[]
  for index,lyris in enumerate(df['list_lyris']):
    list_lyris_re.append(preprocess_kor_eng_sentence(lyris))
  df['list_lyris'] = list_lyris_re
  return df


import pandas as pd
# 가사 한글 전처리
df_re= df_title_lyris_preprocessing(df_title_lyris2)
#df_re.to_csv('/content/drive/MyDrive/project/NLP_project/크롤링/df_title_lyris_0_1000_re.csv', encoding='utf-8-sig')

# '준비중' 뺌
df_re1=df_re.drop(index=df_re[df_re['list_lyris']=='준비중'].index)
# unnamed 지우기
df_re1.drop(['Unnamed: 0'],axis=1,inplace=True)
# reset_index 만들기
df_re1.reset_index(drop=True,inplace=True)
# df_re1 csv로 만들기
df_re1.to_csv('/content/drive/MyDrive/nlp_project/전처리/최종/melon_3000_4000_re.csv', encoding='utf-8-sig')
```
